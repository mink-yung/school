## Q1. 리스트와 while문을 이용한 짝수/홀수 분류

**[문제]**

다음은 숫자 리스트에서 짝수와 홀수를 분류하여 각각 새 리스트에 추가하는 프로그램입니다. 
빈칸을 채워 코드를 완성하세요.

**[실행 결과]**
```text
짝수: [14, 8, 42, 56, 60]
홀수: [7, 23, 35, 19, 31]
짝수 5개, 홀수 5개
```

**[문제 코드]**

````python
numbers = [14, 7, 23, 8, 42, 35, 19, 56, 31, 60]
even = []
odd = []
i = 0
while i < len(numbers):
    if numbers[i] % 2 == 0:
        even.append(numbers[i])
    else:
        odd.append(numbers[i])
    i += 1

print(f"짝수: {even}")
print(f"홀수: {odd}")
print(f"짝수 {len(even)}개, 홀수 {len(odd)}개")
# YOUR CODE HERE
````

## Q2. 별표(*)로 특정 형태 만들기 (이중 for문 활용)

**[문제]**

이중 for문을 사용하여 오른쪽으로 정렬된 10줄 짜리 삼각형 모양의 별표를 출력하는 프로그램입니다.
코드를 완성해 보세요.

**[실행 결과]**

```text
         * 
        ** 
       *** 
      **** 
     ***** 
    ****** 
   ******* 
  ******** 
 ********* 
**********
```

**[문제 코드]**



````python
# YOUR CODE HERE
for i in range(1, 11):
    for j in range(10-i):
        print(' ', end="")
    for k in range(i):
        print('*', end="")
    print()
````

## Q3. 리스트를 이용한 영어단어 퀴즈

**[문제]**

다음은 리스트를 이용하여 영어 단어 스펠링 맞추기 퀴즈를 만드는 프로그램입니다. 
두 개의 리스트(`questions`, `answers`)를 사용하여 코드를 완성하세요.

**[실행 결과]**

```text
tr_in 에서 밑줄(_) 안에 들어갈 알파벳은?a 
정답입니다! 
b_s 에서 밑줄(_) 안에 들어갈 알파벳은?u 
정답입니다! 
_axi 에서 밑줄(_) 안에 들어갈 알파벳은?h 
틀렸습니다! 
air_lane 에서 밑줄(_) 안에 들어갈 알파벳은?p 
정답입니다!
```

**[문제 코드]**


````python
questions = ['tr_in', 'b_s', '_axi', 'air_lane']
answers   = ['a', 'u', 't','p']

# YOUR CODE HERE
for i in range(len(questions)):
    ans = input(f"{questions[i]}에서 밑줄(_) 안에 들어갈 알파벳은? ")
    if ans == answers[i]:
        print("정답입니다!")
    else:
        print("틀렸습니다!")
````

## Q4. while문에서 리스트 활용하기

**[문제]**

다음은 `while` 반복문과 리스트와 조건문을 사용하여 100점 만점의 점수에 대해 학생들의 성적을 A, B, C, D, F 등급으로 나누어 각 등급에 몇 명이 있는지 세어 보는 프로그램입니다. 코드를 완성해 보세요.

**[실행 결과]**

```text
A : 3명 
B : 6명 
C : 3명 
D : 4명 
F : 4명
```

**[문제 코드]**


````python
s = [64, 89, 100, 85, 77, 58, 79, 67, 96, 87,
     87, 36, 82, 98, 84, 76, 63, 69, 53, 22]
count_A = 0           # 90점 이상 ~ 100점 이하
count_B = 0           # 80점 이상 ~ 90점 미만
count_C = 0           # 70점 이상 ~ 80점 미만
count_D = 0           # 60점 이상 ~ 70점 미만
count_F = 0           # 60점 미만

# YOUR CODE HERE
i = 0
while i < len(s):
    if s[i] >= 90:
        count_A += 1
    elif s[i] >= 80:
        count_B += 1
    elif s[i] >= 70:
        count_C += 1
    elif s[i] >= 60:
        count_D += 1
    else:
        count_F += 1
    i += 1


print(f'A : {count_A}명')
print(f'B : {count_B}명')
print(f'C : {count_C}명')
print(f'D : {count_D}명')
print(f'F : {count_F}명')
````

## Q5. while문과 리스트를 이용한 합계/평균 구하기

**[문제]**

다음은 키보드로 성적을 입력받아 리스트에 저장한 다음, 합계와 평균을 구하는 프로그램입니다.  
`while True` 무한 루프와 `if` 조건문을 사용하여 사용자가 -1을 입력할 때까지 성적을 계속 입력받아 리스트 (scores) 에 저장한 후, 전체 합계와 평균을 계산하도록 코드를 작성하세요.

**[실행 결과]**

```text
성적을 입력하세요(종료 시 -1 입력): 75 
성적을 입력하세요(종료 시 -1 입력): 83 
성적을 입력하세요(종료 시 -1 입력): 88 
성적을 입력하세요(종료 시 -1 입력): 93 
성적을 입력하세요(종료 시 -1 입력): 97 
성적을 입력하세요(종료 시 -1 입력): -1 
합계 : 436, 평균 : 87.20
```

**[문제 코드]**


````python
# YOUR CODE HERE
scores = []
while True:
    score = int(input("점수를 입력하세요 (종료하려면 -1 입력): "))
    if score == -1:
        break
    else:
        scores.append(score)
    
sum = 0
for i in range(len(scores)):
    sum += scores[i]
avg = sum / len(scores)
print(f"합계: {sum}, 평균: {avg:.2f}")
````
