```python
x = int(input("Enter a number: "))

if x > 0:
    print("The number is positive.")
else:
    print("The number is not positive or 0.")
```

```python
age = int(input("Enter your age: "))
pay = '3000 won'

if age >= 65:
    pay = 'free'

print(f"The entrance fee is {pay}.")
```

```python
height = int(input("Enter your height in cm: "))
msg = "You are tall enough to ride the roller coaster."
if height < 120:
    msg = "You are not tall enough to ride the roller coaster."

print(msg)
```
