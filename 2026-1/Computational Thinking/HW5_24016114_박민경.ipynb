삼각형의 밑변(width), 높이(height)를 입력받아서 삼각형의 면적을 구하는 `cal_area()`라는 함수를 정의하세요.  
또한 사용자로부터 삼각형의 밑변과 높이를 입력 받은 후, 이 함수를 이용하여 결과를 얻는 프로그램을 완성해 보세요.

```
높이를 입력하시오: 15

```

### Q5. 함수와 리스트
3개 숫자를 오름차순으로 정렬하여 출력하는 `sort3()` 함수를 구현하세요.  
(힌트: 함수 안에서 리스트를 만들고 `append()` 메소드로 세 숫자를 순서대로 추가한 후, `sort()` 메소드를 이용해 정렬하고 리스트를 반환하세요.)  
사용자로부터 3개의 숫자를 입력받고, 이들 숫자를 오름차순으로 출력하는 프로그램을 완성해 보세요.

#### 실행 결과
```
세 수를 입력하세요:
3
2
5
정렬된 리스트는 다음과 같습니다: [2, 3, 5]
```

````python
def sort3(a, b, c):
  num_list = []
  num_list.append(a)
  num_list.append(b)
  num_list.append(c)
  num_list.sort()
  return num_list

print('세 수를 입력하세요:')
n1 = int(input())
n2 = int(input())
n3 = int(input())

sorted_list = sort3(n1, n2, n3)
print('정렬된 리스트는 다음과 같습니다:', sorted_list)
````

### Q6. 함수와 딕셔너리
사용자가 입력한 단어와 정답을 비교하여 '맞습니다!' 또는 '틀렸습니다!' 메시지를 반환하는 `matchWord()` 함수를 만들고,  
이 함수를 이용해 딕셔너리에 있는 모든 단어에 대한 퀴즈를 내는 프로그램을 완성하세요.

#### 실행 결과
```
오렌지에 맞는 영어 단어는? orange
맞습니다!
과자에 맞는 영어 단어는? cookie
맞습니다!
어머니에 맞는 영어 단어는? motheer
틀렸습니다!
...
```

````python
def matchWord(word, answer):
    if word == answer:
        msg = '맞습니다!'
    else:
        msg = '틀렸습니다!'
    return msg

eng_dict = {'orange':'오렌지', 'cookie':'과자', 'mother':'어머니', 'brother':'형제', 'python':'파이썬'}

for key in eng_dict:
    string = input(eng_dict[key] + '에 맞는 영어 단어는? ')
    result = matchWord(string, key)
    print(result)
````

### Q8. 최대공약수 알고리즘 구현
두 수를 입력받아 최대공약수를 계산하는 함수 `computeMaxGong()`을 완성해 보세요. 최대공약수는 두 수를 모두 나누어 떨어지게 하는 가장 큰 정수입니다.

#### 실행 결과
```
첫 번째 수를 입력하세요: 33
두 번째 수를 입력하세요: 44
33과(와) 44의 최대공약수 : 11
```

````python
def computeMaxGong(x,y):
    if x > y:
        small = y
    else:
        small = x
        
    for i in range(1, small + 1):
        if ((x % i == 0) and (y % i == 0)):
            result = i
    return result

num1 = int(input("첫 번째 수를 입력하세요: "))
num2 = int(input("두 번째 수를 입력하세요: "))

max_gong = computeMaxGong(num1, num2)

print(f'{num1}과(와) {num2}의 최대공약수 : {max_gong}')
````

### Q9. 파일 입출력
아래 `scores.txt` 파일의 내용을 읽어 각 학생의 이름, 점수 합계, 평균을 계산하여 출력하는 프로그램을 완성하세요.  
강의 슬라이드 32페이지에서 생성된 파일과 동일합니다.  
※ 파일이 없다면, 아래 파일 내용을 복사하여 `scores.txt` 라는 이름으로 직접 만들어 (현재 ipynb 파일과 같은 폴더에 위치하도록 저장해야함) 실습을 진행하세요.

**scores.txt 파일 내용:**
```
김소영 82 80 93 97 93 88
정예린 86 100 93 86 90 77
이세영 91 88 99 79 92 68
정수정 86 100 93 89 92 93
박지수 80 100 95 89 90 84
```


#### 실행 결과
```
--------------------------------------------------
김소영
합계 : 533, 평균 : 88.83
--------------------------------------------------
정예린
합계 : 532, 평균 : 88.67
--------------------------------------------------
...
```


````python
# scores.txt 파일을 먼저 생성하고 아래 코드를 실행하세요.
file = open('scores.txt', 'r', encoding='utf8')
lines = file.readlines()
file.close()

print('-' * 50)
for line in lines:
    student = line.split()
    i = 0
    sum = 0
    
    while i < len(student): 
        if i == 0:
            print(student[0]) # 학생 이름 출력
        else:
            sum += int(student[i])
        i += 1
        
    avg = sum / 6
    print(f'합계 : {sum}, 평균 : {avg:.2f}')
    print('-' * 50)
````
