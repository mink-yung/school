## 문제 1: 삼각형의 면적 구하기
다음은 삼각형의 밑변의 길이와 높이를 이용하여 삼각형의 면적을 구하는 프로그램입니다. 빈칸(①, ②, ③...)을 채워 전체 코드를 완성하세요.

**[실행 결과]**
```text
삼각형의 밑변 길이 : 10
삼각형의 높이 : 3
삼각형의 면적 : 15.0
```

**[문제 코드]**
</div>

````python
# YOUR CODE HERE
width = 10
height = 3
area = width*height/2
print('삼각현의 밑변 길이:', width)
print('삼각형의 높이:',height)
print('삼각형의 면적:',area)
````

## 문제 2: 성적 평균 구하기
학생 이름과 3과목 성적을 입력받아 총점과 평균을 구하는 프로그램입니다. 소수점 첫째 자리에서 반올림하여 출력하도록 빈칸을 채우세요.

**[실행 결과]**
```text
학생 이름을 입력하세요 : 황예린
국어 성적을 입력하세요 : 83
영어 성적을 입력하세요 : 85
수학 성적을 입력하세요 : 91
이름:황예린, 국어:83, 영어:85, 수학:91, 평균:86.3점
```

**[문제 코드]**
</div>

````python
# YOUR CODE HERE
name = input('학생 이름을 입력하세요: ')
kor = int(input('국어 성적을 입력하세요: '))
eng = int(input('영어 성적을 입력하세요: '))
math = int(input('수학 성적을 입력하세요: ')) 

sums = kor + eng + math
avg = sums / 3
print(f'이름:{name}, 국어:{kor}, 영어:{eng}, 수학:{math}, 평균{avg:.1f}점')
````
