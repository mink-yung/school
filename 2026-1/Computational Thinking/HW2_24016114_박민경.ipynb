## Q1. 세 수 중 큰수 찾기

세 수 중 큰수 찾기 프로그램을 작성하세요.

**[실행 결과]**

```text
첫 번째 정수를 입력하세요 : 34
두 번째 정수를 입력하세요 : 33
세 번째 정수를 입력하세요 : 27
입력된 세 수 34, 33, 27 중에서 가장 큰 수는 34 입니다.
```

**[문제 코드]**

````python
# YOUR CODE HERE
num1 = int(input("첫번째 정수를 입력하세요 : "))
num2 = int(input("두번째 정수를 입력하세요 : "))
num3 = int(input("세번째 정수를 입력하세요 : "))
if num1 > num2 and num1 > num3:
    largest = num1
elif num2 > num1 and num2 > num3:
    largest = num2
else:    
    largest = num3

print(f"입력된 세 수 {num1}, {num2}, {num3} 중 가장 큰 수는 {largest}입니다.")
````

## Q2. 교내 장학금 자격 판별

교내 장학금을 신청하기 위해서는 직전 학기 성적(gpa)과 어학 점수(toeic)가 모두 일정 기준을 넘어야 합니다.
* 성적 기준: 3.5점 이상
* 어학 기준: 800점 이상

사용자로부터 성적(변수명: gpa) 과 어학 점수(변수명: toeic) 를 각각 입력 받고, 위 두 조건을 모두 만족해야 "신청 가능" 을 출력하고, 아니면 "신청 불가" 를 출력하는 프로그램을 구현해보세요.

**[실행 결과]**

```text
직전 학기 성적을 입력하세요: 4.0
어학 점수를 입력하세요: 790
신청 불가
```

````python
# YOUR CODE HERE
gpa = float(input("직전 학기 성적을 입력하세요 : "))
toeic = int(input("어학 점수를 입력하세요 : "))
if gpa >= 3.5 and toeic >= 800:
    print("신청 가능")
else:
    print("신청 불가")
````

## Q3. 미세먼지 등급 판별 문제

사용자로부터 미세먼지 농도(변수명: PM10)를 입력받아, 아래 기준에 따라
공기 질 등급을 결정하고, 그에 따라 알맞은 메시지를 출력하는 프로그램을 작성하세요.

| 등급 | 농도 범위 ($\mu g/m^3$) | 출력 메시지 |
| :--- | :--- | :--- |
| 좋음 | 30 이하 | "현재 공기가 맑습니다. 산책하기 좋아요!" |
| 보통 | 31 ~ 80 | "공기 질이 보통입니다." |
| 나쁨 | 81 ~ 150 | "공기가 나쁩니다. 마스크를 착용하세요." |
| 매우 나쁨 | 151 이상 | "실외 활동을 자제하고 실내에 머무르세요." |

**[실행 결과]**

```text
미세먼지 농도를 입력하세요: 80
공기 질이 보통입니다.
```

**[실행 결과]**

```text
미세먼지 농도를 입력하세요: 81
공기가 나쁩니다. 마스크를 착용하세요.
```

````python
# YOUR CODE HERE
PM10 = int(input("미세먼지 농도를 입력하세요 : "))
if PM10 <= 30:
    print("현재 공기가 맑습니다. 산책하기 좋아요!")
elif PM10 <= 80:
    print("공기 질이 보통입니다.")
elif PM10 <= 150:
    print("공기가 나쁩니다. 마스크를 착용하세요")
else:
    print("실외 활동을 자제하고 실내에 머무르세요.")  
````

## Q4. 5의 배수가 아닌 수의 합계 구하기

5의 배수가 아닌 수의 합계 구하기 프로그램을 작성하세요.

**[실행 결과]**

```text
시작 수를 입력하세요 : 100
끝 수를 입력하세요 : 200
-------------------------------------------------
100에서 200까지 5의 배수가 아닌 수의 합계 : 12000
```

**[문제 코드]**

````python
# YOUR CODE HERE
n1 = int(input("시작 수를 입력하세요 : "))
n2 = int(input("끝 수를 입력하세요 : "))
sum = 0
for i in range(n1, n2 + 1):
    if i % 5 != 0:
        sum += i

print('-'* 50)
print(f'{n1}부터 {n2}까지 5의 배수가 아닌 수의 합: {sum}')
````

하이픈(-)을 포함한 휴대폰 번호를 입력하세요: 010-1234-5678
01012345678

## Q6. 전화번호에 하이픈(-) 추가하기

전화번호에 하이픈(-) 추가하기 프로그램을 작성하세요.

**[실행 결과]**

```text
하이픈(-)을 뺀 11자리의 휴대폰 번호를 입력하세요: 01033334444
010-3333-4444
```

**[문제 코드]**

````python
# YOUR CODE HERE
phone = input('하이픈(-)을 뺀 전화번호를 입력하세요 : ')
number = ''
for i in range(0,11):
    if i == 2:
        number = number +(phone[2]+'-')
    elif i == 6:
        number = number +(phone[6]+'-')
    else:
        number = number + phone[i]

print(number)
````

## Q7. 홀수 개수 세기

홀수 개수 세기 프로그램을 작성하세요.

**[실행 결과]**

```text
숫자를 입력하세요 : 3049894
입력된 숫자 중 홀수의 개수 : 3개
```

**[문제 코드]**

````python
# YOUR CODE HERE
number = input("숫자를 입력하세요 : ")
total = 0
for a in number:
    a = int(a)
    if a % 2 == 1:
        total += 1

print(f"입력된 숫자 중 홀수의 개수: {total}개")
````
