```python
i=200
count=0
while i<601:
    if i%3!=0:
        print(i,end=" ")
        count+=1
    
    if count%8==0:
        print()

    i+=1
```

```python
n1=int(input("enter the first number: "))
n2=int(input("enter the second number: "))
n = int(input("enter the number of terms: "))
sum=0
i = n1
while i<=n2+1:
    if i%n==0:
        sum+=i
    i+=1
print(f"Sum of multiples of {n} between {n1} and {n2}: {sum}")
```

```python
sentence=input('write a sentence: ')
i = len(sentence)-1
while i>=0:
    if sentence[i]=='':
        print('-',end="")
    else:
        print(sentence[i],end="")

    i-=1
```
