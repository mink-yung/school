```python
scores = [[96, 84, 80], [96, 86, 76], [76, 95, 83], [89, 96, 69]]
for i in range(len(scores)):
    print('학생', i+1, '의 점수는', end=' ')
    for j in range(len(scores[i])):
        print(scores[i][j], end=' ')
    print()
```

```python
scores = [[96, 84, 80], [96, 86, 76], [76, 95, 83], [89, 96, 69]]
for i in range(len(scores)):
    sum = 0
    for j in range(len(scores[i])):
        sum += scores[i][j]

    avg = sum / len(scores[i])
    print(f'{i+1}번째 학생의 합계: {sum}, 평균: {avg:.2f}')
```

```python
scores = {'김예진': 90, '박영진': 95, '김소희': 84}

sum = 0
for key in scores:
    sum += scores[key]
    print(f'{key}: {scores[key]}')

avg = sum/len(scores)

print(f"합계:{sum},평균:{avg:.2f}")
```
