# 문제 3

url의 적합성을 체크하는 함수를 만드시오.
1. url 요청이 성공 했다면, 'success', url, status code를 출력하시오.
1. url 요청이 실패하면 'fail'과 url, status code를 출력하시오.

***실행결과***
```
success: https://www.naver.com/, 200
failed: https://www.naver.com/a, 404
```

````python
import requests as rq
def check_url(url):
    res = rq.get(url)
    if res.status_code == 200:
        print(f"success : {url},{res.status_code}")
    else:
        print(f"failed : {url},{res.status_code}")

## 아래는 수정하지 마시오.
check_url('https://www.naver.com/')
check_url('https://www.naver.com/a')
````

| 단계 | 코드 표현 | 데이터 타입 및 특징 | 비고 |
| :--- | :--- | :--- | :--- |
| **요청 및 수신** | `res = rq.get(url)` | **`Response` 객체** | 서버의 응답 전체를 담고 있는 복합 객체 |

# 문제 5

요청해더(request header)을 출력하는 함수 print_request_header() 를 작성하시오.

참고)
| 단계 | 코드 표현 | 데이터 타입 및 특징 | 비고 |
| :--- | :--- | :--- | :--- |
| **객체 연결** | `res.request` | **`PreparedRequest` 객체** | 보낸 요청의 URL, 헤더, 바디 정보를 담은 별도의 객체 |
| **요청 헤더 접근** | `res.request.headers` | **`CaseInsensitiveDict`** | 파이썬 requests 라이브러리가 자동으로 생성한 기본 헤더 정보 |
| **항목 추출** | `.items()` | **`ItemsView` (튜플 쌍)** | `(Key, Value)` 쌍을 반복문에서 사용하기 위한 필수 메서드 |
| **포맷 출력** | `f"{key}: {value}"` | **`f-string`** | 요청 헤더의 이름과 값을 보기 좋게 결합하는 문자열 기법 |


### 결과
```
[구글 요청헤더]
User-Agent: python-requests/2.32.3
Accept-Encoding: gzip, deflate, br
Accept: */*
Connection: keep-alive

[다음 요청헤더]
User-Agent: python-requests/2.32.3
Accept-Encoding: gzip, deflate, br
Accept: */*
Connection: keep-alive
```

````python
import requests as rq
def print_request_header(url):
    res = rq.get(url)
    for key, value in res.request.headers.items():
        print(f"{key} : {value}")
    pass


#아래는 수정하지 마시오.
print("[구글 요청헤더]")
print_request_header('https://www.google.com')
print()
print("[다음 요청헤더]")
print_request_header('https://www.daum.net')
````
