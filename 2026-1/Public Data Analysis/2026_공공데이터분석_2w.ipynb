## 실습3

```python
fig=plt.figure(figsize=(7,7), dpi=100)
ax=fig.subplots()

data = pd.read_table('xy_price.txt', sep='\t')
data.head()

def f1(x):
    if x>=20:
        col='r'
    else:
        col='y'
    return col

data['color']=data['price'].map(f1)
#data['color']=data['price'].map(lambda x: 'r' if x>=20 else 'y')
data.head()
ax.scatter(data['x'], data['y'], marker='P', color=data['color'], edgecolor='k', s=500)
```
